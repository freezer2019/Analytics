```python
def van_guenut(h, alpha=0.0230, theta_r=0.0946, theta_s=0.4674, n=1.2089):
  calcul1 = h/(1*9.81)
  calcul2 = (((calcul1*alpha)**n + 1)**(-1 + 1/n))*(theta_s - theta_r) + theta_r
  return calcul2
```
